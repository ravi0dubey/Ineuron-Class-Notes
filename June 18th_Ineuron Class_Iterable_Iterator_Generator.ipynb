# Generator

```python

#Square of each element till range of element given as input

def sqr_numb(a):
    l = []
    for i in range (1,a+1):
        l.append(i*i)
    return l

sqr_numb(7)
```

```python
#Square of each element till range of element given as input

def sqr_numb1(a):
    l = [i*i for i in range(1,a+1)]
    return l

sqr_numb(7)
```

```python
#range is a geneartor function which does not give element of output directly 

range(1,7)
```

```python
type(range(1,7))
```

```python
# to get output of range function we need to use for loop

for i in range(1,7):
    print(i)
```

```python
#Generator Function

# Instead of using return statment which returns the output directly
# here yield keyword is used to store the output

def sqr_numb1(a):
    l = [i*i for i in range(1,a+1)]
    yield l

#calling function by passing parameter will not return any value
sqr_numb1(7)
```

```python
#to get output from function we need to use for loop  

for i in sqr_numb1(7):
    print(i)
```

```python
#to get output from function we need to use next statement

next(sqr_numb1(5))
```

# Fibonacci series

```python
def fib_series(n):
    l = []
    a = 1
    b = 1
    l.append(a)
    l.append(b)
    for i in range(3,n+1):
        c = a + b
        a = b
        b = c
        l.append(c)
                
    return (l)

fib_series(10)
```

```python
def fib_seriesa(n):
    l = []
    a = 1
    b = 1
    for i in range(1,n+1):
        l.append(a)
        a,b= b, a+b
        
    return (l)

fib_seriesa(10)
    
```

```python
# Generator fucntion in fibonnacci series

def fib_seriesy(n):
    l = []
    a = 1
    b = 1
    for i in range(1,n+1):
        yield a
        a,b= b, a+b
        
    
fib_seriesy(10)
```

```python
# Fetching value of Fibonnacci generator 

for i in fib_seriesy(50):
    print(i)
```
