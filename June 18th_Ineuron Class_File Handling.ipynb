# Setting Pointer

```python
#setting pointer at 0th location of file
f3.seek(0)
```

```python
#reading value starting for 0th position
f3.read()
```

```python
#if we perform read again it will give blank as all data was already read
f3.read()
```

```python
#setting pointer at 1st position
f3.seek(1)
```

```python
#reading value starting for 1st position so we see character T was truncated
f3.read()
```

```python
#setting pointer at 20th position
f3.seek(20)
```

```python
#reading value starting for 20th position
f3.read()
```

# Tell Function

#tell will let us  know where your pointer is currently

```python
#tell will let us  know where your pointer is currently
```

```python
f4= open("file5.txt","r")
```

```python
# when we open file, pointer is by default at 0th location
f4.tell()
```

```python
f4.read()
```

```python
#after read operations pointer is at last position
f4.tell()
```

```python
#setting pointer at 15th position
f4.seek(15)
```

```python
#it returns output as 15
f4.tell()
```

```python
# read operations is performed from 15th position
f4.read()
```

```python
#setting pointer at 21st position
f4.seek(21)
```

```python
#if we pass parameer in read funciton it will read n characters where n is the parameter
#reading only 4 characters from the poistion where pointer is set
f4.read(4)
```
