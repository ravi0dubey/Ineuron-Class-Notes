# Map Function

MAP iterates through each element of parameter and map it to function

```python
l = [2,3,4,5,5,6]

def sq_function(l):
    l1= []
    for i in l:
        l1.append(i*i)
    return l1

a= sq_function(l)
a
```

```python
def sq_function1(n):
    return n*n
```

```python
l2 = [2,3,4,5,5,6]

list(map(sq_function1,l2))
```

```python
list(map(lambda x: x*x,l2))
```

```python
def str_c(n):
    return str(n)
list(map(str_c,l2))
```

```python
list(map(str,l2))
```

```python

l3 = [3,4,5,6,7,8,9]

# a = 3
# b = 4
# a = a+b = 3+4 => 7

# a = sum of a+b = 7
# b = next element of list l3 which is 5

# a = a + b , 7 +5  => 12
# b = next element of list l3 which is 6

# a =  a + b , 12 + 6 => 18
# b = next element of list l3 which is 7

reduce(lambda a,b:a+b,l3)
```

```python
#it will not take 3 parameters

reduce(lambda a,b,c:a+b+c,l3)
```

```python
# When we give only 1 parameter it will take a = 3 , b = 1 by default

l4= [3]
reduce(lambda a,b:a+b,l4)
```

# Filter

it filter out dataset at any point of time

it will work on true or false return value

```python
#filter even number form the list

l5= [1,2,3,4,45,5,98,78]
l6 = []
for i in l5:
    if i%2 == 0:
        l6.append(i)
print(l6)
```

```python
list(filter(lambda x: x%2==0,l5))
```

```python
def check_even(list1):
    l6 = []
    for i in list1:
        if i%2 == 0:
            l6.append(i)
    return(l6)
```

```python
def check_even1(list1):
    if list1%2 == 0:
        return True
```
